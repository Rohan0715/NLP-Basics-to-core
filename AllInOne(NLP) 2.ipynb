```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.corpus import wordnet as wn

# Define the words
word1 = "dog"
word2 = "cat"

# Find the first synset for each word
synset1 = wn.synsets(word1)[0]
synset2 = wn.synsets(word2)[0]

# Print the synsets
print(f"Synset of {word1}: {synset1}")
print(f"Synset of {word2}: {synset2}")

# Calculate path similarity
path_similarity = synset1.path_similarity(synset2)
print(f"Path similarity between {word1} and {word2}: {path_similarity}")

# Calculate Wu-Palmer similarity
wup_similarity = synset1.wup_similarity(synset2)
print(f"Wu-Palmer similarity between {word1} and {word2}: {wup_similarity}")
```

```python
import nltk
from nltk.corpus import wordnet

# Ensure you have the necessary resources
# nltk.download('wordnet')

# Select a word to analyze
word = "car"

# Get all synsets for the word
synsets = wordnet.synsets(word)

# Display relationships for each synset
for synset in synsets:
    print(f"Synset: {synset.name()}")
    print(f"Definition: {synset.definition()}")
    
    # Hypernyms
    hypernyms = synset.hypernyms()
    if hypernyms:
        print("\tHypernyms:")
        for hypernym in hypernyms:
            print(f"\t\t{hypernym.name()} - {hypernym.definition()}")
    
    # Hyponyms
    hyponyms = synset.hyponyms()
    if hyponyms:
        print("\tHyponyms:")
        for hyponym in hyponyms:
            print(f"\t\t{hyponym.name()} - {hyponym.definition()}")
    
    # Holonyms
    holonyms = synset.member_holonyms() + synset.part_holonyms() + synset.substance_holonyms()
    if holonyms:
        print("\tHolonyms:")
        for holonym in holonyms:
            print(f"\t\t{holonym.name()} - {holonym.definition()}")
    
    # Meronyms
    meronyms = synset.member_meronyms() + synset.part_meronyms() + synset.substance_meronyms()
    if meronyms:
        print("\tMeronyms:")
        for meronym in meronyms:
            print(f"\t\t{meronym.name()} - {meronym.definition()}")
    
    print("\n")
```
